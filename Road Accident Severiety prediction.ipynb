```python
# # storing categorical column names to a new variable
categorical=[i for i in df.columns if df[i].dtype=='O']
print('The categorical variables are',categorical)
```

```python
#importing label encoing module for feature selection 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
#create new data frame for performing chi2 test
df1=pd.DataFrame()
```

```python
#add all categorical column to new data frame
for i in categorical:
    if i!= 'Accident_severity':
        df1[i]=le.fit_transform(df[i])
```

```python
df1.head()
```

```python
#importing  chi2 test
from sklearn.feature_selection import chi2
values=chi2(df1,df['Accident_severity'])
values
```

```python
#add this to a new data frame for easy access
f_p_value=pd.DataFrame({'features':df1.columns, 'Fscore': values[0], 'Pvalues':values[1]})
f_p_value
```

we want high Fscore and low Pvalues, 
so we have to remove columns like Owner_of_vehicle, Type_of_vehicle, Road_surface_conditions, Pedestrian_movement,Casualty_severity,Educational_level,Day_of_week,Sex_of_driver,Road_allignment, Sex_of_casualty

```python
df1=df.drop(['Owner_of_vehicle', 'Type_of_vehicle', 'Road_surface_conditions', 'Pedestrian_movement','Casualty_severity','Educational_level','Day_of_week','Sex_of_driver','Road_allignment','Sex_of_casualty'],axis=1)
df1
```

convet string into numerical data using get dummies

```python
#get_dummies
dummy=pd.get_dummies(df1[['Age_band_of_driver', 'Vehicle_driver_relation', 'Driving_experience',
                          'Area_accident_occured', 'Lanes_or_Medians', 'Types_of_Junction', 'Road_surface_type', 
                          'Light_conditions', 'Weather_conditions', 'Type_of_collision', 'Vehicle_movement', 
                          'Casualty_class', 'Age_band_of_casualty', 'Cause_of_accident']],drop_first=True)
dummy.head()
```

```python
#combine dummy df and old df
df2=pd.concat([df1,dummy],axis=1)
df2
```

```python
#drop dummied column
df3=df2.drop(['Age_band_of_driver', 'Vehicle_driver_relation', 'Driving_experience',
                          'Area_accident_occured', 'Lanes_or_Medians', 'Types_of_Junction', 'Road_surface_type', 
                          'Light_conditions', 'Weather_conditions', 'Type_of_collision', 'Vehicle_movement', 
                          'Casualty_class', 'Age_band_of_casualty', 'Cause_of_accident'],axis=1)
df3
```

# Model Creation

KNeighborsClassifier (KNN)

```python
#KNN Algorithm
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train,y_train)
```

```python
y_pred_KNN=classifier.predict(x_test)
y_pred_KNN
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,ConfusionMatrixDisplay
print(confusion_matrix(y_test,y_pred_KNN))
label=['Slight Injury','Serious Injury','Fatal injury' ]
result1=confusion_matrix(y_test,y_pred_KNN)
result1
cmd=ConfusionMatrixDisplay(result1,display_labels=label)
cmd.plot()
```

```python
report1=classification_report(y_test,y_pred_KNN)
print(report1)
```

```python
#accuracy score 
score=accuracy_score(y_test,y_pred_KNN)
score
```
